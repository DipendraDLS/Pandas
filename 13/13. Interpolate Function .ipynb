## Interpolate Function ---> Used to fill NaN values

```python
import pandas as pd
```

```python
df = pd.read_csv('/home/dipendra/django/Pandas_Practice/student_10.csv')
df
```

**1. Guesses the integer value and fill it into NaN values**

```python
df.interpolate()
```

**2. method = time --> Interpolates accoirding to time but some dataset column has to be convert into Timestamp as shown below**

```python
df.interpolate(method= 'time')
```

***As you can see that type is of str so we need to convert the type using parse_dates parameter***

```python
type(df.Date[0])
```

```python
df = pd.read_csv('/home/dipendra/django/Pandas_Practice/student_10.csv', parse_dates = ['Date'])
df
```

```python
type(df.Date[0])
```

***Changing index to date column***

```python
df = pd.read_csv('/home/dipendra/django/Pandas_Practice/student_10.csv', parse_dates = ['Date'], index_col = 'Date')
df
```

***Interpolate according to the Date column***

```python
df.interpolate(method= 'time')
```

**3. method = 'index' --> interpolate according to the index column.. Here our index column is changed to Date**

```python
df.interpolate(method = 'index')
```

***Changing index column to 'class' column*** 

```python
df = pd.read_csv('/home/dipendra/django/Pandas_Practice/student_10.csv', parse_dates = ['Date'], index_col = 'Class')
df
```

```python
df.interpolate(method='index')
```

**4. method = 'nearest' fills with nearest upper column value**

```python
df.interpolate(method='nearest')
```

**5.method = 'polynomial' and we have to specify order of polynomial too.**

```python
df = pd.read_csv('/home/dipendra/django/Pandas_Practice/student_10.csv', parse_dates = ['Date'], index_col = 'Date')
df
```

```python
df.interpolate(method = 'polynomial', order =1)
```

```python
df.interpolate(method = 'polynomial', order =2)
```

**6.method = 'spline' and we have to specify order of polynomial too.**

```python
df.interpolate(method = 'spline', order =2)
```
