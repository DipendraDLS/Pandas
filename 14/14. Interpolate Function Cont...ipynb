```python
df.interpolate()
```

**1. Limit ---> Specifies how many consecutive
Nan values to be filled**

***Bydefault limit takes method = 'linear'***

```python
df
```

```python
df.interpolate(limit =1)
```

```python
df.interpolate(limit =2)
```

**2. limit_direction ---> Specifies how should(backward/forward) consecutive Nan values to be filled**

```python
df
```

***fills the Nan Value from bottom in each column***

```python
df.interpolate(limit =1, limit_direction='backward')
```

```python
df.interpolate(limit =2, limit_direction='backward')
```

***fills one NaN value at forward and one NaN value at backward***

```python
df.interpolate(limit =1, limit_direction='both')
```

**3. limit_area ---> Specifies how should Nan values to be filled i.e (inside(interpolate)/outside(extrapolate))**

```python
df.interpolate(limit_area = 'inside')
```

***Doesn't fills the NaN values if it's outside the valid values***

```python
df.interpolate(limit_area = 'outside')
```

**4. inplace --> Updates the original dataset values**

```python
df.interpolate(limit_area = 'outside')
```
