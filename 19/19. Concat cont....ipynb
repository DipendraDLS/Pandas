```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class':[5,6,7,8] })
df1
```

```python
df2 = pd.DataFrame({'ID':[3,4],
              'Name':['C','D'],
              'Class':[7,8] })
df2
```

```python
pd.concat([df1,df2])
```

**1. Concat Column wise**

```python
pd.concat([df1,df2], axis = 1)
```

**2. Join parameter ---> inner-> joins the common index only. outer--> joins all the dataframe values**

```python
pd.concat([df1,df2], axis = 1, join='inner')
```

```python
pd.concat([df1,df2], axis = 1, join='outer')
```

**3. Keys parameter ---> used to seperate the multiple dataframe by givin the labels**

```python
pd.concat([df1,df2], keys=['First df', 'Second df'])
```

```python
pd.concat([df1,df2], axis = 1, keys=['First df', 'Second df'] )
```

**4. To concat different dataframe with different column names**

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
              'Name':['A','B','C','D'],
              'Class':[5,6,7,8] })
df1
```

```python
df3 = pd.DataFrame({'Marks':[40,63,91,34]})
df3
```

```python
pd.concat([df1,df3])
```
